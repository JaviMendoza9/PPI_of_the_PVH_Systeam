```python
import re
import pandas as pd
import glob

# Lista de archivos que coinciden con el patrón "TriTrypDB-i_Tcruzi*.fasta"
archivos_fasta = glob.glob("TriTryDB/TriTrypDB-*_Tcruzi*.fasta")

# Lista para almacenar los datos extraídos
genes_lista = []

# Expresión regular para extraer el valor de gene=
patron = r"\| gene=([^|]+) \|"

# Procesar cada archivo
for archivo in archivos_fasta:
    with open(archivo, "r") as f:
        contenido = f.read()
    
    # Buscar todas las coincidencias en el archivo
    genes_encontrados = re.findall(patron, contenido)
    
    # Agregar los genes a la lista
    genes_lista.extend(genes_encontrados)

# Crear un DataFrame y guardar en un .tsv
df_genes = pd.DataFrame({"Gene_Names_ORF": genes_lista})
df_genes.to_csv("TriTryDB/genes_extraidos_TriTrypDB_AnnotatedProteins.tsv", sep="\t", index=False)
```

```python
# Cargar el primer archivo (contiene Protein_ID)
file1 = "Combined_Proteins_DB/combined_Proteins_TriTrypDB.tsv"
df1 = pd.read_csv(file1, sep="\t")

# Cargar el segundo archivo (contiene preferred_name)
file2 = "string_data/353153.protein.info.v12.0.txt"
df2 = pd.read_csv(file2, sep="\t")

# Verificar si la columna existe en ambos archivos
if "Protein_ID" in df1.columns and "preferred_name" in df2.columns:
    # Filtrar los valores de Protein_ID que están en preferred_name
    coincidencias = df1[df1["Protein_ID"].isin(df2["preferred_name"])]

    # Imprimir coincidencias encontradas
    if not coincidencias.empty:
        print("Se encontraron coincidencias:")
        print(coincidencias)
    else:
        print("No se encontraron coincidencias.")
else:
    print("Error: La columna 'Protein_ID' o 'preferred_name' no existe en los archivos.")
```
