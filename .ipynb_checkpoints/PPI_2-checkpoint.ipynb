```python
# Crear el grafo
G = nx.Graph(name="Protein Interaction Graph")

# Convertir DataFrame a numpy array para procesamiento rápido
interactions = df_filtered.to_numpy()

# Diccionario para mapear nodos con su ncbiTaxonId
node_taxon_mapping = {}

# Iterar eficientemente sobre el array numpy
for a, b, w, taxon_id in interactions[:, :4]:  
    try:
        # Convertir peso a float
        w = float(w)

        # Manejar Taxon_ID con múltiples valores (toma el primero)
        if isinstance(taxon_id, str):
            taxon_ids = list(map(int, taxon_id.split(","))) if "," in taxon_id else [int(taxon_id)]
        else:
            taxon_ids = [int(taxon_id)]
        
        # Asignar el primer Taxon_ID a los nodos (solo si aún no están en el diccionario)
        node_taxon_mapping.setdefault(a, taxon_ids[0])
        node_taxon_mapping.setdefault(b, taxon_ids[0])

        # Agregar arista con peso al grafo
        G.add_edge(a, b, weight=w)

    except ValueError as e:
        print(f"Error en la conversión de datos en {a, b, w, taxon_id}: {e}")

# Asignar atributo ncbiTaxonId a los nodos
nx.set_node_attributes(G, node_taxon_mapping, "ncbiTaxonId")
```

```python
# Calculate the number of nodes in the graph
num_nodes = nx.number_of_nodes(G)

# Calculate the number of edges in the graph
num_edges = nx.number_of_edges(G)

# Print the number of proteins involved
print("Number of proteins involved:", num_nodes)

# Print the number of interactions
print("Number of interactions:", num_edges)
```

```python
def unique_position_layout(G, scale=1, max_attempts=1000):
    np.random.seed(42)  # Para reproducibilidad
    occupied_positions = set()
    positions = {}

    for node in G.nodes():
        for _ in range(max_attempts):
            x, y = np.random.uniform(-scale, scale), np.random.uniform(-scale, scale)
            pos = (round(x, 3), round(y, 3))  # Redondeo para evitar precisión excesiva

            if pos not in occupied_positions:
                positions[node] = pos
                occupied_positions.add(pos)
                break
        else:
            print(f"Advertencia: No se encontró una posición libre para el nodo {node}")
    return positions
```

```python
# Mapeo de especies a coordenada Z más separada
taxon_to_z = {13249: -10, 9606: 10, 353153: 30}  # Más separados
color_dict = {9606: 'blue', 13249: 'red', 353153: 'green'}

# Diccionario de colores más claros para las aristas dentro de la misma especie
light_color_dict = {9606: "lightblue", 13249: "lightcoral", 353153: "lightgreen"}

marker_dict = {
    9606: 'd',  # Triángulo
    13249: 's',  # Cuadrado
    353153: 'o'   # Círculo
}

# Obtener posiciones en 2D
pos_2d = unique_position_layout(G, scale=1)  # Asegurar posiciones únicas

# Escalar X e Y
scale_factor = 100
pos_3d = {node: (x * scale_factor, y * scale_factor, taxon_to_z[node_taxon_mapping[node]]) 
    for node, (x, y) in pos_2d.items()}

# Obtener atributos ncbiTaxonId 
node_taxon_mapping = nx.get_node_attributes(G, "ncbiTaxonId")
```

```python
# Crear figura 3D
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# Dibujar nodos con diferentes formas según la especie
for node, (x, y, z) in pos_3d.items():
    taxon_id = node_taxon_mapping.get(node, None)
    node_color = color_dict.get(taxon_id, 'gray')  # Color del nodo
    node_marker = marker_dict.get(taxon_id, 'o')  # Forma del nodo
    ax.scatter(x, y, z, color=node_color, s=25, marker=node_marker, edgecolor='black', linewidth=0.8)  # Contorno negro

# Contar nodos por especie
node_counts = {taxon_id: sum(1 for n in G.nodes if node_taxon_mapping.get(n) == taxon_id) for taxon_id in color_dict.keys()}

# Conectar puntos dentro de la misma especie
def conectar_puntos_3d(nodes, z_value, ax, is_same_species=True):
    puntos_2d = np.array([pos_3d[n][:2] for n in nodes])  # Solo X, Y
    tree = KDTree(puntos_2d)

    for i, node in enumerate(nodes):
        _, vecinos = tree.query(puntos_2d[i], k=3)  # 3 vecinos más cercanos
        taxon_id = node_taxon_mapping.get(node)
        edge_color = light_color_dict.get(taxon_id, 'gray')  # Obtener el color claro

        for j in vecinos[1:]:
            vecino = nodes[j]
            ax.plot([pos_3d[node][0], pos_3d[vecino][0]], 
                    [pos_3d[node][1], pos_3d[vecino][1]], 
                    [z_value, z_value], color=edge_color, alpha=0.8, linewidth=1.5)

# Conectar nodos por especie (misma especie)
for taxon_id, z_value in taxon_to_z.items():
    nodos_especie = [n for n in G.nodes if node_taxon_mapping.get(n) == taxon_id]
    conectar_puntos_3d(nodos_especie, z_value, ax, is_same_species=True)

# Conectar interacciones entre especies (diferentes especies)
for a, b in G.edges:
    if a in pos_3d and b in pos_3d:
        z1, z2 = pos_3d[a][2], pos_3d[b][2]
        if z1 != z2:  # Solo si están en diferentes planos (entre especies)
            ax.plot([pos_3d[a][0], pos_3d[b][0]], 
                    [pos_3d[a][1], pos_3d[b][1]], 
                    [z1, z2], color='gray', alpha=0.5, linestyle='dashed', linewidth=0.1)

# Crear bordes con opacidad variable
def dibujar_bordes(ax, z_value, alpha_value=0.3):
    # Definir las esquinas del plano
    x_lim = [-scale_factor, scale_factor]
    y_lim = [-scale_factor, scale_factor]
    
    # Crear los bordes del plano conectando las esquinas sin conectarlas entre sí
    for x1 in x_lim:
        ax.plot([x1, x1], [y_lim[0], y_lim[1]], [z_value, z_value], color='black', lw=1, alpha=alpha_value)
    for y1 in y_lim:
        ax.plot([x_lim[0], x_lim[1]], [y1, y1], [z_value, z_value], color='black', lw=1, alpha=alpha_value)

# Ajuste de transparencia para crear la proyección
for z_value in taxon_to_z.values():
    # Ajustar la opacidad (alpha) de manera que los planos más atrás tengan menos opacidad
    alpha_value = 0.3 if z_value == 20 else 0.5  # Menos opacidad para los planos más lejanos
    dibujar_bordes(ax, z_value, alpha_value)

# Ajustar límites
ax.set_xlim(-scale_factor, scale_factor)
ax.set_ylim(-scale_factor, scale_factor)
ax.set_zlim(-5, 25)  # Ajustar el límite Z para reducir el espacio entre el título y los planos

# Ocultar ejes y malla
ax.grid(False)
ax.set_axis_off()

# Añadir leyenda
plt.title("PPI Network of the PVH System", pad=20)  # Ajuste del espacio para el título
legend_labels = [
    plt.Line2D([0], [0], marker=marker_dict.get(tax_id, 'o'), color="w", markerfacecolor=color, markersize=10, 
               markeredgecolor='black', label=f"{tax_id} ({node_counts.get(tax_id, 0)} nodes)")
    for tax_id, color in color_dict.items()
]
plt.legend(handles=legend_labels, title="ncbiTaxonId", loc="upper right")

# Guardar la imagen
plt.savefig("1_Fig.png", format="png", dpi=300, bbox_inches="tight")

# Mostrar gráfico
plt.show()
```

# 1. Connectivity Analysis

## 1.1. Network Diameter

```python
# Diccionario de asignación de Taxon_ID
TAXON_MAPPING = {
    "_HUMAN": 9606,
    "_RHOPR": 13249,
    "_TRYCC": 353153
}

def obtener_taxon_id(protein_id):
    """Asigna el Taxon_ID basado en la terminación del Protein_ID"""
    for suffix, taxon in TAXON_MAPPING.items():
        if protein_id.endswith(suffix):
            return taxon
    return "Desconocido"  # Si no coincide con ninguna especie

def contar_especies(G):
    """Cuenta la cantidad de nodos por especie en un subgrafo"""
    taxon_counts = Counter(obtener_taxon_id(node) for node in G.nodes)
    return taxon_counts

def calcular_diametros_por_especie(G):
    """Calcula el diámetro promedio para cada especie considerando componentes conexas dentro de cada una."""
    especies = set(obtener_taxon_id(n) for n in G.nodes)
    diametros = {}

    for especie in especies:
        nodos_especie = [n for n in G.nodes if obtener_taxon_id(n) == especie]
        subgraph = G.subgraph(nodos_especie).copy()

        if len(subgraph) <= 1:
            diametros[especie] = "No calculable"
            continue

        # Separar en componentes conexas dentro de la especie
        subcomponentes = list(nx.connected_components(subgraph))
        diametros_especie = []

        for subcomp in subcomponentes:
            sub_subgraph = subgraph.subgraph(subcomp)
            if len(sub_subgraph) > 1 and nx.is_connected(sub_subgraph):
                diametros_especie.append(nx.diameter(sub_subgraph))

        # Calcular el diámetro promedio de las subredes de la especie
        if diametros_especie:
            diametros[especie] = sum(diametros_especie) / len(diametros_especie)
        else:
            diametros[especie] = "No calculable"

    return diametros

def analizar_componentes(G):
    """Analiza la conectividad, calcula diámetros y guarda los resultados en un archivo TSV."""
    
    results = []
    output_path = "Results/Connectivity_Analysis/S9_Table.tsv"

    if nx.is_connected(G):
        print("La red es conexa, no se segmentará en componentes.")
        return
    
    components = list(nx.connected_components(G))
    print(f"La red no es conexa. Tiene {len(components)} componentes.")
    
    for i, component in enumerate(components):
        subgraph = G.subgraph(component).copy()

        # Calcular diámetros
        diametros_componente = nx.diameter(subgraph) if nx.is_connected(subgraph) and len(subgraph) > 1 else "No calculable"
        diametros_especie = calcular_diametros_por_especie(subgraph)  # ← ¡Aquí se usa la nueva función!

        # Filtrar valores válidos para max, min y promedio
        diametros_validos = [d for d in diametros_especie.values() if isinstance(d, (int, float))]

        # Estadísticas de diámetro con manejo de "No calculable"
        diametros_max = max(diametros_validos, default="No calculable")
        diametros_min = min(diametros_validos, default="No calculable")
        diametro_promedio = sum(diametros_validos) / len(diametros_validos) if diametros_validos else "No calculable" Imprimir detalles de la componente
        print(f"\nComponente {i+1}: {len(component)} nodos")
        print(f"  Diámetro de la componente ≈ {diametros_componente}")
        print(f"  Diámetro promedio ≈ {diametro_promedio}")
        print(f"  Diámetro máximo: {diametros_max}")
        print(f"  Diámetro mínimo: {diametros_min}")
            
        for taxon_id, count in contar_especies(subgraph).items():
            print(f"  Especie {taxon_id}: diámetro promedio ≈ {diametros_especie.get(taxon_id, 'No calculable')}, {count} nodos")

        #
        
        # Generar todas las combinaciones de pares de nodos dentro de la componente
        nodes = list(subgraph.nodes)
        for j in range(len(nodes)):
            for k in range(j + 1, len(nodes)):
                protein1 = nodes[j]
                protein2 = nodes[k]
                
                # Asignar Taxon_IDs correctamente
                taxon1 = obtener_taxon_id(protein1)
                taxon2 = obtener_taxon_id(protein2)

                # Obtener el camino más corto entre ambos nodos
                try:
                    path = nx.shortest_path(subgraph, source=protein1, target=protein2)
                    path_str = ",".join(path)
                    path_length = len(path) - 1  # Diámetro real del camino
                except nx.NetworkXNoPath:
                    path_str = "No hay camino"
                    path_length = "No calculable"

                results.append([protein1, taxon1, protein2, taxon2, i+1, path_length, path_str])

    # Guardar los resultados en un archivo TSV con ordenamiento
    df = pd.DataFrame(results, columns=["Protein_ID1", "Taxon_ID1", "Protein_ID2", "Taxon_ID2", "Componente", "Diametro", "Path"])

    # Ordenar por: Componente -> Diametro (desc) -> Misma especie (desc) -> Protein_ID1 -> Protein_ID2
    df["EsMismaEspecie"] = (df["Taxon_ID1"] == df["Taxon_ID2"]).astype(int)  # 1 si son de la misma especie, 0 si son de diferentes especies

    df = df.sort_values(by=["Componente", "Diametro", "EsMismaEspecie", "Protein_ID1", "Protein_ID2"], ascending=[True, False, False, True, True])

    df = df.drop(columns=["EsMismaEspecie"])  # Eliminar columna auxiliar
    
    df.to_csv(output_path, sep="\t", index=False)
```

```python
# Ejecutar análisis en la red G
analizar_componentes(G)
```

## 1.2. Degree

```python
# Función para calcular estadísticas de grado
def degree_statistics(G):
    degrees = [G.degree(n) for n in G.nodes()]
    avg_degree = np.mean(degrees)
    max_degree = np.max(degrees)
    min_degree = np.min(degrees)
    return avg_degree, max_degree, min_degree

# Función para calcular estadísticas de grado por especie en un componente
def degree_statistics_by_species(G):
    species_degree = {}
    for node in G.nodes:
        species_id = G.nodes[node].get('ncbiTaxonId')
        degree = G.degree(node)
        if species_id not in species_degree:
            species_degree[species_id] = []
        species_degree[species_id].append(degree)
    
    species_avg_degree = {}
    for species_id, degrees in species_degree.items():
        species_avg_degree[species_id] = np.mean(degrees)
    
    return species_avg_degree, species_degree

# Verificar si la red es conexa o no
if nx.is_connected(G):
    # Si la red es conexa, calcular grado en toda la red
    avg_degree, max_degree, min_degree = degree_statistics(G)
    print("La red es conexa. Tiene 1 componente.")
    print(f"  Grado promedio: {avg_degree:.2f}")
    print(f"  Grado máximo: {max_degree}")
    print(f"  Grado mínimo: {min_degree}")
    
else:
    # Si la red no es conexa, calcular grado en cada componente
    components = list(nx.connected_components(G))
    print(f"La red no es conexa. Tiene {len(components)} componentes.")
    
    for i, component in enumerate(components):
        subgraph = G.subgraph(component)
        avg_degree, max_degree, min_degree = degree_statistics(subgraph)
        print(f"\nComponente {i+1}: {len(component)} nodos")
        print(f"  Grado promedio ≈ {avg_degree:.2f}")
        print(f"  Grado máximo: {max_degree}")
        print(f"  Grado mínimo: {min_degree}")
        
        # Calcular y mostrar grado por especie
        species_avg_degree, species_degree = degree_statistics_by_species(subgraph)
        species_count = {9606: "Human", 13249: "R. prolixus", 353153: "T. cruzi"}
        for species_id, avg_deg in species_avg_degree.items():
            print(f"  Especie {species_count.get(species_id, 'Desconocida')} ({species_id}): Grado promedio ≈ {avg_deg:.2f}")
```

```python
def edge_weight_statistics(G):
    """ Calcula estadísticas de peso de las aristas de un grafo """
    edge_weights = [G[edge[0]][edge[1]]['weight'] for edge in G.edges()]
    avg_weight = np.mean(edge_weights) if edge_weights else 0
    max_weight = np.max(edge_weights) if edge_weights else 0
    min_weight = np.min(edge_weights) if edge_weights else 0
    return avg_weight, max_weight, min_weight
```

```python
def species_statistics_by_edges(G):
    """ Calcula la cantidad de nodos y el peso promedio de las aristas por especie """
    species_data = {}
    
    for edge in G.edges:
        u, v = edge
        species_ids = {G.nodes[u].get('ncbiTaxonId'), G.nodes[v].get('ncbiTaxonId')}
        weight = G[u][v]['weight']
        
        for species_id in species_ids:
            if species_id:
                if species_id not in species_data:
                    species_data[species_id] = {"count": 0, "weights": []}
                species_data[species_id]["count"] += 1
                species_data[species_id]["weights"].append(weight)

    species_stats = {
        sp: {
            "count": data["count"],
            "avg_weight": np.mean(data["weights"])
        }
        for sp, data in species_data.items()
    }
    
    return species_stats
```

```python
def interaction_level(G):
    """ Calcula el nivel de interacción y la energía total entre enlaces """
    interaction = G.number_of_edges()
    total_energy = sum(weight for u, v, weight in G.edges.data('weight'))
    
    return interaction, total_energy

def edge_weights(G):
    """ Obtiene el peso de las aristas """
    return {edge: G[edge[0]][edge[1]]['weight'] for edge in G.edges}
```

```python
def analizar_componentes_por_peso(G):
    """ Analiza la conectividad y estadísticas de peso de las aristas en la red """
    species_count = {9606: "Human", 13249: "R. prolixus", 353153: "T. cruzi"}

    if nx.is_connected(G):
        avg_weight, max_weight, min_weight = edge_weight_statistics(G)
        print("La red es conexa. Tiene 1 componente.")
        print(f"  Peso promedio: {avg_weight:.2f}")
        print(f"  Peso máximo: {max_weight}")
        print(f"  Peso mínimo: {min_weight}")

        species_stats = species_statistics_by_edges(G)
        if len(species_stats) > 1:
            for species_id, stats in species_stats.items():
                print(f"  Especie {species_count.get(species_id, 'Desconocida')} ({species_id}): {stats['count']} aristas, Peso promedio ≈ {stats['avg_weight']:.2f}")
        else:
            species_id, stats = list(species_stats.items())[0]
            print(f"  Especie {species_count.get(species_id, 'Desconocida')} ({species_id}): Peso promedio ≈ {stats['avg_weight']:.2f}")

        interaction, total_energy = interaction_level(G)
        print(f"Nivel de interacción: {interaction}")
        print(f"Total de energía entre enlaces: {total_energy}")
    else:
        components = list(nx.connected_components(G))
        print(f"La red no es conexa. Tiene {len(components)} componentes.")

        for i, component in enumerate(components):
            subgraph = G.subgraph(component).copy()
            avg_weight, max_weight, min_weight = edge_weight_statistics(subgraph)

            print(f"\nComponente {i+1}: {len(component)} aristas")
            print(f"  Peso promedio: {avg_weight:.2f}")
            print(f"  Peso máximo: {max_weight}")
            print(f"  Peso mínimo: {min_weight}")

            species_stats = species_statistics_by_edges(subgraph)
            if len(species_stats) > 1:
                for species_id, stats in species_stats.items():
                    print(f"  Especie {species_count.get(species_id, 'Desconocida')} ({species_id}): {stats['count']} aristas, Peso promedio ≈ {stats['avg_weight']:.2f}")
            else:
                species_id, stats = list(species_stats.items())[0]
                print(f"  Especie {species_count.get(species_id, 'Desconocida')} ({species_id}): Peso promedio ≈ {stats['avg_weight']:.2f}")

            # Mostrar nivel de interacción y energía total por componente
            interaction, total_energy = interaction_level(subgraph)
            print(f"Nivel de interacción: {interaction}")
            print(f"Total de energía entre enlaces: {total_energy}")
```

```python
def calcular_pesos_y_rutas(G):
    results = []
    total_nodos = len(G.nodes)
    # Iterar sobre cada nodo y calcular caminos más cortos solo con nodos alcanzables
    for i, protein1 in enumerate(G.nodes, start=1):
        lengths, paths = nx.single_source_dijkstra(G, source=protein1, weight='weight')
        for protein2, path in paths.items():
            if protein1 != protein2:  # Evitar caminos a sí mismo
                path_weight = lengths[protein2]
                path_str = ",".join(path)
                results.append([protein1, protein2, path_weight, path_str])
        
    # Convertir a DataFrame
    df = pd.DataFrame(results, columns=["Protein1", "Protein2", "Weight", "Shortest_Paths"])
    df = df.sort_values(by=["Weight", "Protein1", "Protein2"], ascending=[False, True, True])
    
    # Guardar como archivo .tsv
    output_path = "Results/proteins/S15_Table.tsv"
    df.to_csv(output_path, sep="\t", index=False)
```

```python
random.seed(1)

# Convertir el grafo de NetworkX a iGraph
def nx_to_igraph(G):
    """ Convierte un grafo de NetworkX a iGraph conservando las aristas y pesos. """
    nodes = list(G.nodes)
    edges = list(G.edges)
    weights = [G[u][v].get("weight", 1.0) for u, v in edges]

    g = ig.Graph()
    g.add_vertices(len(nodes))
    g.add_edges([(nodes.index(u), nodes.index(v)) for u, v in edges])
    g.es["weight"] = weights
    return g, nodes

# Convertir NetworkX a iGraph
G_igraph, nx_nodes = nx_to_igraph(G)

# Detección de comunidades con CNM en iGraph
dendrogram_cnm = G_igraph.community_fastgreedy(weights="weight")
optimal_count_cnm = dendrogram_cnm.optimal_count
clusters_cnm = dendrogram_cnm.as_clustering()
membership_cnm = clusters_cnm.membership
modularity_cnm = clusters_cnm.q

print(f"CNM Optimum community count: {optimal_count_cnm}")
print(f"Modularity: {modularity_cnm:.4f}")

# Convertir la estructura de comunidades a listas de nodos
communities = [[] for _ in range(optimal_count_cnm)]
node_to_community = {}
for i, node in enumerate(nx_nodes):
    communities[membership_cnm[i]].append(node)
    node_to_community[node] = membership_cnm[i]

# Función para generar posiciones únicas sin superposición dentro de cada comunidad
def generate_positions(G, communities, spacing=7, inner_radius=2):
    """ Genera posiciones únicas sin superposición dentro de cada comunidad con forma aleatoria. """
    np.random.seed(42)
    positions = {}

    angle_step = 2 * np.pi / len(communities)
    community_centers = {
        i: (np.cos(i * angle_step) * spacing, np.sin(i * angle_step) * spacing)
        for i in range(len(communities))
    }

    for i, community in enumerate(communities):
        cx, cy = community_centers[i]
        num_nodes = len(community)

        angles = np.linspace(0, 2 * np.pi, num_nodes, endpoint=False) + np.random.uniform(-0.3, 0.3, num_nodes)
        radii = np.random.uniform(0.5, inner_radius, num_nodes)

        for j, node in enumerate(community):
            x = cx + radii[j] * np.cos(angles[j])
            y = cy + radii[j] * np.sin(angles[j])
            positions[node] = (x, y)

    return positions

pos = generate_positions(G, communities, spacing=7, inner_radius=2)

# Crear la figura
plt.figure(figsize=(12, 10))

# Usar un colormap claro para las comunidades
cmap = plt.cm.coolwarm  
colors = [cmap(i / (len(communities) - 1)) for i in range(len(communities))]

# Convertir colores a tonos más claros
def lighten_color(color, factor=0.6):
    """ Aclara el color mezclándolo con blanco. """
    return tuple(np.array(color[:3]) * factor + np.array([1, 1, 1]) * (1 - factor)) + (1,)

community_colors = {i: lighten_color(colors[i]) for i in range(len(communities))}

# Dibujar edges entre comunidades
for a, b in G.edges:
    if a in pos and b in pos:
        comm_a, comm_b = node_to_community[a], node_to_community[b]
        if comm_a != comm_b:
            plt.plot([pos[a][0], pos[b][0]], 
                     [pos[a][1], pos[b][1]], 
                     color="gray", alpha=0.2, linestyle="dashed", linewidth=0.1)

# Dibujar edges dentro de la misma comunidad con borde negro
for a, b in G.edges:
    if a in pos and b in pos:
        comm_a, comm_b = node_to_community[a], node_to_community[b]
        if comm_a == comm_b:
            edge_color = community_colors[comm_a]
            plt.plot([pos[a][0], pos[b][0]], 
                     [pos[a][1], pos[b][1]], 
                     color=edge_color, alpha=0.6, linewidth=0.7)
            plt.plot([pos[a][0], pos[b][0]], 
                     [pos[a][1], pos[b][1]], 
                     color="black", alpha=1.0, linewidth=0.2)

# Dibujar nodos
for i, community in enumerate(communities):
    nodes = list(community)
    color = community_colors[i]
    nx.draw_networkx_nodes(G, pos, nodelist=nodes, 
                           node_color=[color], node_size=25, alpha=0.9, 
                           edgecolors="black", linewidths=0.8)

    # Dibujar círculo alrededor de la comunidad
    x_vals, y_vals = zip(*[pos[n] for n in nodes])
    center_x, center_y = np.mean(x_vals), np.mean(y_vals)
    radius = max(np.linalg.norm(np.array([x, y]) - np.array([center_x, center_y])) for x, y in zip(x_vals, y_vals))
    circle = plt.Circle((center_x, center_y), radius * 1.5, color=color, fill=False, linewidth=2, linestyle="dashed", alpha=0.6)
    plt.gca().add_patch(circle)

# Crear leyenda con cantidad de nodos por comunidad
legend_elements = [
    Line2D(
        [0], [0], marker='o', color='w',
        label=f'Community {i} ({len(community)} nodes)',
        markerfacecolor=community_colors[i], markersize=10
    )
    for i, community in enumerate(communities)
]
plt.legend(handles=legend_elements, loc='upper left')

# Configuración final
plt.title('Communities of Algorithm CNM of PPI Network of the VPH System')
plt.axis("off")

# Guardar la imagen
plt.savefig('Results/Community_Detection/2_Figure.png', format='png', dpi=300, bbox_inches='tight')

plt.show()
```
